##### Raw Dataset Location

```python
import pandas as pd
import numpy as np
```

```python
from pathlib import Path
DATA_PATH = Path() / "../data"
DATA_PATH.mkdir(parents=True,exist_ok=True)
```

```python
def load_csv(filename, data_path=DATA_PATH,encoding='ISO-8859-1'):
    csv_path = data_path / filename
    return pd.read_csv(csv_path,encoding=encoding)

def save_csv(data, filename, data_path=DATA_PATH,encoding='ISO-8859-1'):
    csv_path = data_path / filename
    data.to_csv(csv_path, index=False,encoding='ISO-8859-1')
```

```python
supplyChain = load_csv("DataCoSupplyChainDataset.csv")
```

```python
supplyChain.info()
```

```python
print(supplyChain.dtypes)
```

```python
supplyChain['order date (DateOrders)'] = pd.to_datetime(supplyChain['order date (DateOrders)'])
```

```python
supplyChain = supplyChain.sort_values(by='order date (DateOrders)', ascending=True).reset_index(drop=True)
```

```python
start_date = '2015-01-01'
end_date = '2015-03-31'

# Filter the DataFrame
Q1_2015 = supplyChain[(supplyChain['order date (DateOrders)'] >= start_date) & (supplyChain['order date (DateOrders)'] <= end_date)].reset_index(drop=True)
Q1_2015
```

```python
save_csv(Q1_2015, 'Q1_2015.csv')
```

```python
future_data = supplyChain[supplyChain['order date (DateOrders)'] > end_date].reset_index(drop=True)
future_data
```

```python
save_csv(future_data, 'future_data.csv')
```
