#### Demand Forecasting

#### LSTM Time Series

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dropout, Dense
```

```python
def create_sequences(data, n_steps):
    X, y = [], []
    for i in range(len(data)):
        # find the end of the current sequence
        end_ix = i + n_steps
        # check if we are beyond the dataset
        if end_ix > len(data)-1:
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = data[i:end_ix, :-1], data[end_ix, -1]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
df = X_selected.copy()
df["Order Item Quantity"] = supplyChain_df["Order Item Quantity"]
dataset = df.to_numpy()
n_features = dataset.shape[1] - 1  # number of features
n_steps = 5  # number of time steps

# create sequences
X_lstm, y_lstm = create_sequences(dataset, n_steps)

# X is in the shape [samples, time steps, features]
print(X_lstm.shape)
# y is a 1D array
print(y_lstm.shape)
```

```python
# Get the train and test set
from sklearn.model_selection import train_test_split
X_train_lstm, X_test_lstm, y_train_lstm, y_test_lstm = train_test_split(X_lstm, y_lstm, test_size = 0.20, random_state = 695)
```

```python
X_train_lstm.shape
```

Experiment 1

```python
lstm_model = Sequential([
    Input(shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])),
    LSTM(50, activation='relu'),
    Dropout(0.2),
    Dense(1)
])

lstm_model.summary()
```

```python
lstm_model.compile(optimizer='adam', loss='mse', metrics=['mse'])
epo = 100
history = lstm_model.fit(X_train_lstm, y_train_lstm, epochs=epo, 
                         batch_size=64, 
                         validation_split=0.2,
                         verbose=2)
```

```python
acc = history.history['mse']
val_acc = history.history['val_mse']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epo)

plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training MSE')
plt.plot(epochs_range, val_acc, label='Validation MSE')
plt.legend(loc='lower right')
plt.title('Training and Validation MSE')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
y_pred_lstm = lstm_model.predict(X_test_lstm)
print("R^2:", r2_score(y_test_lstm, y_pred_lstm))
print("MSE:", mean_squared_error(y_test_lstm, y_pred_lstm))
print("MAE:", mean_absolute_error(y_test_lstm, y_pred_lstm))
```

#### Random Forest Regressor

```python
# Get the train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size = 0.20, random_state = 695)
```

```python
import warnings
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Suppress warnings
warnings.filterwarnings('ignore')

# Define model
rfr = RandomForestRegressor(random_state=695)

# Corrected parameter grid
forest_params = [{
    'n_estimators': [30, 50, 70, 90],
    'max_features': [5, 10, 15, 20, 'sqrt', 'log2'],
    'min_samples_split': [2, 50, 100, 150, 200]  #  removed 1
}]
```

```python
# Grid search setup
reg = GridSearchCV(
    rfr,
    forest_params,
    cv=3,
    scoring='neg_mean_squared_error',
    verbose=False,    # set verbose=False to hide fitting messages
    error_score='raise'  # optional: raise instead of warning if errors occur
)

# Fit model
reg.fit(X_train, y_train)
```

```python
# Convert the cv_results_ to a DataFrame
results_df = pd.DataFrame(reg.cv_results_)

# Select relevant columns to display
relevant_columns = ['rank_test_score', 'mean_test_score', 'std_test_score', 'params']

# Display the DataFrame sorted by rank_test_score
display_df = results_df.loc[:, relevant_columns].sort_values('rank_test_score')
print(display_df)
```

```python
print(reg.best_params_)
print(reg.best_score_) # MSE
```

```python
# Test set
from sklearn.metrics import r2_score
y_pred = reg.predict(X_test)
mean_squared_error(y_test, y_pred)
```

```python
# Best model rerun
from sklearn.metrics import r2_score
rfr = RandomForestRegressor(random_state = 695, 
                            max_features = 20, min_samples_split = 50, n_estimators = 90)
rfr.fit(X_train, y_train)
y_pred_rf = rfr.predict(X_test)
print("R^2:", r2_score(y_test, y_pred_rf))
print("MSE:", mean_squared_error(y_test, y_pred_rf))
print("MAE:", mean_absolute_error(y_test, y_pred_rf))
```

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, validation_scores = learning_curve(
    estimator=rfr,
    X=X_train,
    y=y_train,
    train_sizes=np.linspace(0.1, 1.0, 10),
    cv=3, # Cross-validation strategy
    scoring='neg_mean_squared_error', 
    n_jobs=-1 # Use all cores
)
```

```python
train_scores = train_scores *(-1)
validation_scores = validation_scores *(-1)
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
validation_scores_mean = np.mean(validation_scores, axis=1)
validation_scores_std = np.std(validation_scores, axis=1)
```

```python
plt.figure()
plt.title("Random Forest Learning Curve")
plt.xlabel("Training examples")
plt.ylabel("Score")
plt.grid()

plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, color="lightgrey")
plt.fill_between(train_sizes, validation_scores_mean - validation_scores_std,
                 validation_scores_mean + validation_scores_std, color="lightblue")

plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
         label="Training score")
plt.plot(train_sizes, validation_scores_mean, 'o-', color="g",
         label="Cross-validation score")

plt.legend(loc="best")
plt.show()
```
